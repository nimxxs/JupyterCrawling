# 일자별 뉴스기사 크롤링하기
* 기사에 대한 일자는 pdate라는 변수가
* 기사의 각 페이지는 pgi라는 변수에 의해 제어됨
* requests의 get 함수 호출시 params 매개변수에 의해
* 질의문자열의 매개변수를 설정할 수 있음

```python
import requests
from bs4 import BeautifulSoup
import time
```

```python
url = 'https://news.jtbc.co.kr/section/list.aspx'
headers = {'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/115.0.0.0 Safari/537.36'}
```

```python
pdate = input('검색할 기사의 날짜(예:20230814)는? ')
```

```python
titles, previews, categorys, reporters, pdates = [], [], [], [], []
```

```python
pgi = 1

while True:
    params = {'pdate':pdate, 'pgi':pgi}
    res = requests.get(url, headers=headers, params=params)
    html = BeautifulSoup(res.text, 'lxml')

    pages = html.select('div.paginate a.num')
    # print(pages)
    lpage = pages[len(pages) - 1].text # a.num 중 맨 마지막 요소 지정
    # print(pgi, lpage)
    
    print(f'{pgi}번째 페이지 출력중...')

    # 기사제목
    for title in html.select('dt.title_cr a'):
        print(title.text)
        titles.append(title.text)

    pgi += 1
    if int(lpage) < pgi: break
    
    time.sleep(2)
```
