```python
html.prettify()
```

### 지역 입력받아 미세먼지 정보 출력

```python
region = input('지역은? (예:서울 강남구)')
```

```python
for item in html.findAll('item'):
    if (region.split()[0] == item.sidoName.text) and \
        (region.split()[1] == item.stationName.text):
        print(region, item.dataTime.text,
              item.pm10Value.text, item.pm25Value.text)
```

### 미세먼지/초미세먼지 등급에 따른 이모지 출력
* 미세먼지 : ~ 30, ~ 80, ~ 150, 151 ~
* 초미세먼지 : ~ 15, ~ 35, ~ 75, 76 ~
* 😊(좋음) 😗(보통) 🤔(나쁨) 🥵(최악)

```python
def pm10grade(val):
    grade = '😊'
    if val > 150: grade = '🥵'
    elif val > 80: grade = '🤔'
    elif val > 30: grade = '😗'
    return grade
```

```python
def pm25grade(val):
    grade = '😊'
    if val > 75: grade = '🥵'
    elif val > 35: grade = '🤔'
    elif val > 15: grade = '😗'
    return grade
```

```python
for item in html.findAll('item'):
    if (region.split()[0] == item.sidoName.text) and \
        (region.split()[1] == item.stationName.text):
        pm10 = int(item.pm10Value.text)
        pm25 = int(item.pm25Value.text)
        print(region, item.dataTime.text,
              pm10, pm10grade(pm10), pm25, pm25grade(pm25))
```
